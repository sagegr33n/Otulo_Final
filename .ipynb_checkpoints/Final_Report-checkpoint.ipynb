## 1. Analyzing August 2008-09 Chlorophyll Concentration

<!-- In this cell, introduction the data set(s) you are working with, specify which aspect of your research question this section addresses, and describe the figure you will produce. -->


I began with cleaning and plotting the chlorophyll concentration data I collected from neo.gsfc.nasa.gov. This data will help establish my regions of high productivity indicated by higher chlorophyll. I will be producing a figure that will show the August 2008 and 2009 global near surface chlorophyll of the Atlantic Ocean, specifically focusing on regions of high concentration located off the coasts of Africa and South America.

```python
# load data file(s)

# reading in modis data from august 2008 and 2009, dropping unnecessary palette variable
aug_08=xr.open_dataset('data/modis_aug08.nc').drop('palette')
aug_09=xr.open_dataset('data/modis_aug09.nc').drop('palette')
# as part of commenting your code throughout this section, discuss any problems you encountered and how you solved them
```

```python
# perform data cleaning steps

# loading times were long due to global distribution ,
# using coarsen to reduce data granularity 
aug_08=aug_08.coarsen(lat=2,lon=2).mean()
aug_09=aug_09.coarsen(lat=2,lon=2).mean()
```

```python
# perform data analysis

# displaying my clean data to make sure it has all of the variables & dimensions i need
display(aug_08)
display(aug_09)
```

```python
# make and display the first figure

# set up subplots for each monthly chlorophyll
fig, axs=plt.subplots(ncols=2,subplot_kw={'projection':ccrs.PlateCarree()},figsize=(10,12))

# running a loop to set the base extent and features for each plots
for ax in axs.flat:
    ax.set_extent([-90,30,-60,50], crs=ccrs.PlateCarree())
# adding coastlines and gridlines
    ax.add_feature(cfeature.LAND,color='ivory')
    ax.coastlines()
    ax.gridlines(crs=ccrs.PlateCarree(),alpha=0.5)

# plotting the chlorophyll of aug 2008 and aug 2009
# using LogNorm with a min of 0.1 and a max of 1 to best display the differences in
# coastal concentration compared to open ocean regions
pcolormesh1=axs[0].pcolormesh(aug_08['lon'],aug_08['lat'],aug_08['chlor_a'],transform=ccrs.PlateCarree(),
              cmap='viridis', norm=LogNorm(vmin=0.1,vmax=1))
pcolormesh2=axs[1].pcolormesh(aug_09['lon'],aug_09['lat'],aug_09['chlor_a'],transform=ccrs.PlateCarree(),
              cmap='viridis', norm=LogNorm(vmin=0.1,vmax=1))

# add titles
axs[0].set_title('Monthly CHL Concentration (August 2008)',fontsize='12') 
axs[1].set_title('Monthly CHL Concentration (August 2009)',fontsize='12') 

# set colorbar
c=fig.colorbar(pcolormesh1, ax=axs, shrink=0.3,fraction=0.05, pad=0.04)
c.set_label('Chlorophyll (mg/m3)',fontsize='12')
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 1 Caption:** Monthly CHL Concentration of Global Ocean (August 2008-09). Higher coastal concentrations of chlorophyll compared to those of the open ocean can be observed on both figures. This helps prove that there is a general trend on this scale. This will help establish areas of high productivity that may escalate up the food chain and correspond with areas with high foraging rates. 

## 2. Analyzing August 2008-09 Migration Routes of Arctic Terns

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

In this section, I am working with Arctic Tern geolocation data from the Greenland Institute of Natural Resources, specifically analyzing the data collected from birds migrating during August of 2008 and 2009. This data seeks to address the migration route aspect of my research question, and establish a pattern for the connection between high primary productivity areas and regions of higher tracking activity. I will produce a figure showing the geolocation tracks of the tagged birds during their August southward migration in both years. 

```python
# if a new dataset, load data and perform data cleaning steps
# otherwise, continue to data analysis and figure creation

# reading in arctic tern data from august 2008 and aug 2009
arctic_tern=pd.read_csv('data/Arctic_Tern_locations.csv',usecols=['Latitude','Longitude','Mean'])
```

```python
# perform data cleaning 

# converting to pandas df, selecting mean travel times to be datetime referenced
df=pd.DataFrame(arctic_tern)
df['Mean']=pd.to_datetime(df['Mean'])
```

```python
# perform data analysis

# creating mask that selects data collected over span of august in 2008
# assistance from stackxchange on this method
aug08_mask = (df['Mean'] > '2008-08-01 01:29:00') & (df['Mean'] <= '2008-08-31 23:00:00')
# using .loc to create new df of only those values, using copy to avoid warning
aug08_df=df.loc[aug08_mask].copy()

# doing the same for 2009 data
aug09_mask= (df['Mean'] > '2009-08-01 00:35:00') & (df['Mean'] <= '2009-08-31 23:00:00')
aug09_df=df.loc[aug09_mask].copy()
```

```python
# make and display figure
# now plotting my monthly tracking data! using subplots again
fig, axs=plt.subplots(ncols=2,subplot_kw={'projection':ccrs.PlateCarree()},figsize=(10,12))

# running another loop Again to set base extents and projections
for ax in axs.flat:
    ax.set_extent([-90,30,-60,50], crs=ccrs.PlateCarree())
# adding coastlines and gridlines
    ax.add_feature(cfeature.LAND,color='ivory')
    ax.coastlines()
    ax.gridlines(crs=ccrs.PlateCarree(),alpha=0.5)

# using scatter instead of pcolormesh this time b/c they are just single points
axs[0].scatter(aug08_df['Latitude'],aug08_df['Longitude'],color='black',s=5,label='Arctic Tern Geolocation')
axs[1].scatter(aug09_df['Latitude'],aug09_df['Longitude'],color='red',s=5,label='Arctic Tern Geolocation')

# add titles
axs[0].set_title('Arctic Tern Migration (August 2008)',fontsize='12') 
axs[1].set_title('Arctic Tern Migration (August 2009)',fontsize='12') 

# add legends
axs[0].legend()
axs[1].legend()
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 2 Caption:** Arctic Tern Migration (August 2008-2009). From these figures, it can be observed that the Arctic Terns were tracked over these month long periods to travel along the Western Coast of Africa. These figures will help in establishing the most common migration routes of the birds during this time of year, and will help in proving that there is a connection between high areas of productivity and their chosen paths. 

## 3. Analyzing Migration Routes against Chlorophyll Concentration Distribution

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

In this section I will be working with both my Chlorophyll Concentration and Arctic Tern geolocation data and analyzing the resutls of each against eachother. This seeks to prove that there is a connection between the routes and high chlorophyll concentration areas. I will be producing a figure that graphs both the global concentration of chlorophyll and tracking data of the terns during the month of August in 2008 and 2009.

```python
# now to plot on top of eachother.... did not have to perform any more data cleaning!

fig, axs=plt.subplots(ncols=2,subplot_kw={'projection':ccrs.PlateCarree()},figsize=(10,12))

# running another loop to set bases
for ax in axs.flat:
    ax.set_extent([-50,30,-50,40], crs=ccrs.PlateCarree())
    
# adding coastlines and gridlines
    ax.add_feature(cfeature.LAND,color='ivory')
    ax.coastlines()
    ax.gridlines(crs=ccrs.PlateCarree(),alpha=0.5)

# plotting the chlorophyll of aug 2008 and aug 2008
pcolormesh1=axs[0].pcolormesh(aug_08['lon'],aug_08['lat'],aug_08['chlor_a'],transform=ccrs.PlateCarree(),
              cmap='viridis', norm=LogNorm(vmin=0.1,vmax=1))
pcolormesh2=axs[1].pcolormesh(aug_09['lon'],aug_09['lat'],aug_09['chlor_a'],transform=ccrs.PlateCarree(),
              cmap='viridis', norm=LogNorm(vmin=0.1,vmax=1))

c=fig.colorbar(pcolormesh1, ax=axs, shrink=0.3,fraction=0.05, pad=0.04)
c.set_label('Chlorophyll (mg/m3)',fontsize='12')

# plotting the tracking data on top
axs[0].scatter(aug08_df['Latitude'],aug08_df['Longitude'],color='black',s=5,label='Arctic Tern Geolocation')
axs[1].scatter(aug09_df['Latitude'],aug09_df['Longitude'],color='red',s=5,label='Arctic Tern Geolocation')

# add titles
axs[0].set_title('Migration & CHL Concentration (Aug. 2008)',fontsize='12') 
axs[1].set_title('Migration & CHL Concentration (Aug. 2009)',fontsize='12') 

# add legends
axs[0].legend(loc='lower left')
axs[1].legend(loc='lower left')
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 3 Caption:** Arctic Tern Migration & CHL Concentration (August 2008-2009). This figure plots migration data against chlorophyll concentration data to show a correlation between the geolocation tracks of these Arctic Terns and areas of high primary productivity. It can be seen that these birds are tracked over both of these years to migrate close to the coastal waters of Africa, an area that also shows higher concentrations of chlorophyll compared to other regions of the map.
